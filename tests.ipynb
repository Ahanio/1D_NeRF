```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function that returns normal distribution at a point x
def normal(x, mu, sigma):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sigma, 2.))) * 1 / (sigma * np.sqrt(2 * np.pi))
```

```python
# visualize the normal distribution
#plot normal distribution CDF
plt.plot(x, 1 - np.cumsum(normal(x, 5, 1)*(x[1] - x[0])))
plt.show()

x = np.linspace(0, 10, 100)
plt.plot(x, normal(x, 5, 1)*(x[1] - x[0]))
plt.plot(x, x * normal(x, 5, 1)*(x[1] - x[0]))
plt.show()

plt.plot(x, np.cumsum(x * normal(x, 5, 1)*(x[1] - x[0])))
plt.show()

```
